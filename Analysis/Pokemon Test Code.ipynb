```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
```

```python
print(df.head(5))
```

```python
## Read Headers
print(df.columns)
print("\n")

## Read each Column
print(df[["Name", "HP", "Defense"]][0:5])
print("\n")

## Read each Row
print(df.iloc[0:5])
print("\n")

## Read a specific location (Row, Column)
print(df.iloc[2,1])
print("\n")

## Iterate through each Row
for index, row in df.iterrows():
    if index <= 4:
        print(index, row)
        print("\n")
        print(index, row["Name"])
        print("\n")
## Choose rows based on specific Condition
df.loc[df["Type 1"] == "Fire"]
```

```python
## Basic Stats
df.describe()
```

```python
## Sort Values
df.sort_values("Name", ascending=True)
```

```python
## Sort Multiple Criteria
df.sort_values(["Type 1", "HP"], ascending=[1,0])  # Sorts by Type ascending and HP descending
```

```python
## Add column
 # df["Total Stats"]= df["HP"] + df["Attack"] + df["Defence"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
 # df = df.drop(columns=["Total Stats"])

## Different way to add column
df["Total Stats"] = df.iloc[:, 4:10].sum(axis=1) # Goes to 10 as the last column is excluded, so this gives us columns 4 to 9

df.head(5)
```

```python
## Change Columns order
cols = list(df.columns.values)
print(cols)

## Gets the first 4 columns, then gets the last column with -1, then the rest of the columns, and puts them in that order
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] 
df.head(5)
```

```python
## Save as CSV without index column
df.to_csv("Pokemon_Data_Added_Total.csv", index=False)
```

```python
## Filter with "and"
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
## Filter with "or"
df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
```

```python
## Filter multiple "and"
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
```

```python
## Create new df with filtered data
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
print(new_df)
```

```python
## Reset index of df
new_df.reset_index(drop=True, inplace=True)
print(new_df)
```

```python
## Return all rows that contain string in column
df.loc[df["Name"].str.contains("Mega")]
```

```python
## Return all rows that do not contain string in column with ~
df.loc[~df["Name"].str.contains("Mega")]
```

```python
## Regular expressions
import re

## If Or expression
df.loc[df["Type 1"].str.contains("Fire|Grass", regex=True)]
```

```python
## If Or expression ignoring capitalisation
df.loc[df["Type 1"].str.contains("fire|grass", flags=re.I, regex=True)]
```

```python
## Filter based on partial string ( ^ = string must be at start of line , pi = string we are looking for , [a-z] = any letters after our string can be from a-z , * = find all rows that match this condition
df.loc[df["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)]
```

```python
## Conditional Changes
## change string to different name
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Flame"
print(df)
```

```python
## Use condition to set parameter of another column
df.loc[df["Type 1"] == "Flame", "Legendary"] = True # changes all fire types to legendary = true
print(df)
```

```python
## Load a previous checkpoint
df = pd.read_csv("Pokemon_Data_Added_Total.csv")
print(df)
```

```python
## Modify columns based on criteria
df.loc[df["Total Stats"] > 500, ["Generation", "Legendary"]] = ["Test 1", "Test 2"]
print(df)
```

```python
## Aggregate Statistics (Groupby)
## Group together by type, calculate mean, and sort by highest value
df.groupby(["Type 1"]).mean().sort_values("HP", ascending = False)
```

```python
## Sum all column values grouped by type 1
df.groupby(["Type 1"]).sum()
```

```python
## Count all column values grouped by type 1
df.groupby(["Type 1"]).count()
```

```python
## Add count column to assist count function
df["count"] = 1

## Groupby type 1, then type 2, then count
df.groupby(["Type 1", "Type 2"]).count()["count"]
```
